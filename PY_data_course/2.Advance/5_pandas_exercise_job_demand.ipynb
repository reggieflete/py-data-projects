```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_CA = df[df['job_country'] == 'Canada'].copy()
```

```python
df_CA['job_posted_month'] = df_CA['job_posted_date'].dt.strftime('%B')

df_CA
```

```python
df_CA_pivot = df_CA.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_CA_pivot.reset_index(inplace=True)
df_CA_pivot['month_no'] = pd.to_datetime(df_CA_pivot['job_posted_month'], format='%B').dt.month
df_CA_pivot.sort_values('month_no', inplace=True)
df_CA_pivot.set_index('job_posted_month', inplace=True)
df_CA_pivot.drop(columns='month_no', inplace=True)

df_CA_pivot
```

```python
df_CA_pivot.plot(kind='line')
```

```python
top_3 = df_CA['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
```

```python
df_CA_pivot[top_3].plot(kind='line', figsize=(12, 6), title='Top 3 Job Titles in Canada by Month')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```
